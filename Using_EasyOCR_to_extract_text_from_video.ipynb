```python
def extract_city_and_state(input_string):
  input_string = input_string.strip().upper()

  city_pattern = r'([A-Z\s]+)[,;]'
  state_pattern = r'[,;]\s*([A-Z]{2,})'

  city_match = re.search(city_pattern, input_string)
  state_match = re.search(state_pattern, input_string)

  city = city_match.group(1) if city_match else None
  state = state_match.group(1) if state_match else None

  if city is not None and state is not None:
    return city, state

  else:
    return None, None
```

```python
def extract_date_and_time(input_string):
  input_string = input_string.strip().upper()

  # Regular expressions for date and time formats
  date_formats = [
      r"\d{4}-\d{2}-\d{2}",                  # YYYY-MM-DD
      r"\d{2}/\d{2}/\d{4}",                  # MM/DD/YYYY
      r"\d{4}/\d{2}/\d{2}",                  # YYYY/MM/DD
  ]

  time_formats = [
      r"\d{2}\s*:\s*\d{2}\s*:\s*\d{2}",                  # HH:MM:SS
      r"\d{2}\s*:\s*\d{2}\s* [AP]M",                  # HH:MM AM/PM
      r"\d{2}\s*:\s*\d{2}\s*:\s*\d{2}\s* [AP]M",            # HH:MM:SS AM/PM
      r"\d{2}\s*:\s*\d{2}\s*:\s*\d{2}Z",                 # HH:MM:SSZ
  ]

  # Match and extract date formats
  for date_format in date_formats:
      match = re.search(date_format, input_string)
      if match:
        date = match.group(0)
        break
      else:
        date = None

  # Match and extract time formats
  for time_format in time_formats:
      match = re.search(time_format, input_string)
      if match:
        time = match.group(0)
        break
      else:
        time = None

  #if not date is None and not time is None:
  return date, time
  #else:
  #  return None
```

```python
def remove_date_and_time(input_string):
  input_string = input_string.strip().upper()

  # Regular expressions for date and time formats
  date_formats = [
      r"\d{4}-\d{2}-\d{2}",                  # YYYY-MM-DD
      r"\d{2}/\d{2}/\d{4}",                  # MM/DD/YYYY
      r"\d{4}/\d{2}/\d{2}",                  # YYYY/MM/DD
  ]

  time_formats = [
      r"\d{2}\s*:\s*\d{2}\s*:\s*\d{2}",                  # HH:MM:SS
      r"\d{2}\s*:\s*\d{2}\s* [AP]M",                  # HH:MM AM/PM
      r"\d{2}\s*:\s*\d{2}\s*:\s*\d{2}\s* [AP]M",            # HH:MM:SS AM/PM
      r"\d{2}\s*:\s*\d{2}\s*:\s*\d{2}Z",                 # HH:MM:SSZ
  ]

  # Match and extract date formats
  for date_format in date_formats:
    text_unique = re.sub(date_format, "", input_string)

  # Match and extract time formats
  for time_format in time_formats:
    text_unique = re.sub(time_format, "", text_unique)

  return text_unique
```

```python
directory_path = '/content/download_folder'
directory_files = os.listdir(directory_path)
#video_list = ['1.mp4','2.mp4','3.mp4','4.mp4']
for video in directory_files:
  print(video)
  # Create a VideoCapture object and read from input file
  cap = cv2.VideoCapture("/content/download_folder/" + video)

  # Get the frame rate of the video
  fps = cap.get(cv2.CAP_PROP_FPS)

  # Define the desired frame interval (e.g., every second)
  frame_interval = int(fps)  # Extract one frame per second

  # Get the total number of frames in the video
  total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

  # Create a tqdm progress bar
  progress_bar = tqdm(total=total_frames, desc='Processing Frames', unit='frame')

  if (cap.isOpened()== False):
  # Give a error message
    print("Error opening video file")

  data_json = {
      "date": "",
      "start_time" : "",
      "end_time" : "",
      "city" : "",
      "state" : "",
      "description" : "",
      "id": ""
  }
  progress = 0
  frame_count = 0
  final_text1 = ""
  unique_sentences = []
  timeFlag = False
  cityFlag = False
  stateFlag = False
  # Read until video is completed
  while(cap.isOpened()):

  # Capture frame-by-frame
      ret, frame = cap.read()
      if ret == True:
        progress_bar.update(1)
        progress = progress + 1
        if(progress > total_frames):
          break
        frame_count = frame_count + 1
        if frame_count%(frame_interval) == 0:
          cropped_image_location = frame[0:259, 0:1277] # location cropped image
          cropped_image = frame[470:720, 0:1280] # description cropped image
          # Convert the image to a NumPy array
          image_np = np.array(cropped_image)

          # Perform text extraction
          result = reader.readtext(image_np)

          # Convert the image to a NumPy array
          image_np_location = np.array(cropped_image_location)

          # Perform text extraction
          result_location = reader.readtext(image_np_location)

          for detection in result_location:
              text = detection[1]
              city, state = extract_city_and_state(text)
              if city is not None and cityFlag == False:
                data_json["city"] = city
                cityFlag = True
              if state is not None and stateFlag == False:
                data_json["state"] = state
                stateFlag = True

          # Print the extracted text
          for detection in result:
              #print(detection[0])
              #print(detection[1])
              text = detection[1]
              date, time = extract_date_and_time(text)
              if date is not None:
                data_json["date"] = date
              if time is not None and timeFlag == False:
                data_json["start_time"] = time
                timeFlag = True
              elif time is not None and time != "":
                data_json["end_time"] = time
              pattern = r"(?m)^[^\n\s]+\s+.*$"

              matches = re.findall(pattern, text)
              total_text = matches[0] if matches else ""
              #print(total_text)
              total_text = remove_date_and_time(total_text)
              if total_text.strip() != "":
                text_unique = total_text.lower().strip()
                remove_pattern = r"[!#$%&'()*+,-./:;<=>?@[\]^_`{|}~]"
                text_unique = re.sub(remove_pattern, "", text_unique)
                if text_unique not in unique_sentences and (all(text_unique not in sentence for sentence in unique_sentences)) and not any(sentence in text_unique for sentence in unique_sentences):
                  unique_sentences.append(text_unique)
                  final_text1 = final_text1 + " " + total_text
          cv2.waitKey(0)

      # Press Q on keyboard to exit
          if cv2.waitKey(25) & 0xFF == ord('q'):
              break
        #count = count + 1;

  # Break the loop
      else:
          break

  #print("UNIQUE TEXT")
  #print(final_text1)
  data_json["description"] = final_text1
  data_json["id"] = video.split(".")[0];
  print(data_json)

  # Serializing json
  json_object = json.dumps(data_json, indent=4)

  file_name = "/content/download_json/" + video.split(".")[0] + ".json"
  # Writing to sample.json
  with open(file_name, "w") as outfile:
      outfile.write(json_object)
  # When everything done, release
  # the video capture object
  cap.release()

  # Closes all the frames
  cv2.destroyAllWindows()
```
