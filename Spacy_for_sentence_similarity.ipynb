```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_lg")

# Sample accident description (replace with your actual text)
accident_description = """
On a rainy evening, a blue sedan collided with a red SUV on Elm Street in the city center.
The accident occurred around 7:45 PM near the highway exit.
"""

# Process the accident description using spaCy
doc = nlp(accident_description)

# Initialize variables to store extracted information
time_of_day = None
weather_conditions = None
vehicle_types_involved = []
location = None
road_type = None

# Extract information from the parsed text
for token in doc:
    print(token.text)
    if "time" in token.text.lower() and not time_of_day:
        time_of_day = token.text

    if "weather" in token.text.lower() and not weather_conditions:
        weather_conditions = token.text

    if "suv" in token.text.lower() or "sedan" in token.text.lower():
        vehicle_types_involved.append(token.text)

    if "street" in token.text.lower() and not location:
        location = token.text

    if "highway" in token.text.lower() and not road_type:
        road_type = token.text

# Print the extracted information
print("Time of Day:", time_of_day)
print("Weather Conditions:", weather_conditions)
print("Types of Vehicles Involved:", vehicle_types_involved)
print("Location:", location)
print("Road Type:", road_type)
```

```python
#important code from here
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_lg")
```

```python
def DescriptionCleaning(df, columns, name):
  columns_list = []
  columns_list.append(columns)
  columns_list.append(name)
  print(columns_list)
  df = df[columns_list]
  print(df)
  df=df.dropna()

  labelRaw = list(df[columns])
  label = []
  for text in labelRaw:
    text = str(text)
    temp = re.sub('[^a-zA-Z]+', ' ', text)
    temp = temp.lower().strip()
    label.append(temp)

  df[columns]=label

  return df
```

```python
file_path = "/content/Accidents_data_Processed.csv"
df = pd.read_csv(file_path)
df = DescriptionCleaning(df, "Description", "BUSINESS NAME")
```

```python
file_path = "/content/Image to text output.xlsx"
manual_driving_df = pd.read_excel(file_path)
df1 = DescriptionCleaning(manual_driving_df, "Llava output", "Video name")
```

```python
count = 0
data = []
obj = {}
# Creating Empty DataFrame and Storing it in variable df
outputDF = pd.DataFrame()
for i in range(len(df)):
    for j in range(len(df1)):
        obj = {"business_name": "", "prompt1": "", "video_name": "", "prompt2": "", "score": 0}
        row1 = df.iloc[i]
        row2 = df1.iloc[j]
        if i != j:
            prompt1 = row1['Description']
            prompt2 = row2['Llava output']
            s1 = nlp(prompt1)
            s2 = nlp(prompt2)
            sim = s1.similarity(s2)
            obj["business_name"] = row1["BUSINESS NAME"]
            obj["prompt1"] = prompt1
            obj["video_name"] = row2["Video name"]
            obj["prompt2"] = prompt2
            obj["score"] = sim
            data.append(obj)
            if sim > 0.95:
              count = count + 1
              #print(f"Similarity between Prompt {i+1} and Prompt {j+1}: {sim}")
              #print(prompt1)
              #print(prompt2)
              print(i+1)
print(count)
```

```python
outputDF = pd.DataFrame(data)
```
